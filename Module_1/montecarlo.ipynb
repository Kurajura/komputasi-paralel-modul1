```python
import random
import math

def hitung_pi_montecarlo(num_samples):
    inside = 0
    for _ in range(num_samples):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
    return 4 * inside / num_samples

if __name__ == "__main__":
    samples = 10_000_000
    pi = hitung_pi_montecarlo(samples)
    print(f"Estimasi π: {pi}")
    print(f"Error: {abs(pi - math.pi)}")
```
