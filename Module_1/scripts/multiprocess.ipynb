```python
import multiprocessing
import random
import time
import math

def monte_carlo_chunk(chunk_size):
    """Menghitung titik dalam lingkaran untuk sejumlah sampel"""
    inside = 0
    for _ in range(chunk_size):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
    return inside

def calculate_pi_parallel(num_samples, num_processes):
    """Menghitung π menggunakan multiprocessing"""
    chunk_size = num_samples // num_processes
    
    # Buat pool worker
    with multiprocessing.Pool(processes=num_processes) as pool:
        results = pool.map(monte_carlo_chunk, [chunk_size] * num_processes)
    
    total_inside = sum(results)
    pi_estimate = 4 * total_inside / num_samples
    return pi_estimate

# Testing dengan berbagai jumlah proses
if __name__ == "__main__":
    num_samples = 10_000_000
    
    for num_procs in [1, 2, 4, 8, 16, 32]:
        start_time = time.time()
        pi_est = calculate_pi_parallel(num_samples, num_procs)
        end_time = time.time()
        
        print(f"\nProses: {num_procs}")
        print(f"  Estimasi π: {pi_est}")
        print(f"  Waktu: {end_time - start_time:.4f} detik")
```

```python
import multiprocessing
import random
import time
import math

def monte_carlo_chunk(chunk_size):
    inside = 0
    for _ in range(chunk_size):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
    return inside

def calculate_pi_parallel(num_samples, num_processes):
    chunk_size = num_samples // num_processes
    with multiprocessing.Pool(processes=num_processes) as pool:
        results = pool.map(monte_carlo_chunk, [chunk_size] * num_processes)
    total_inside = sum(results)
    return 4 * total_inside / num_samples

if __name__ == "__main__":
    # Variasi ukuran sampel
    ukuran_sampel = [1_000_000, 5_000_000, 10_000_000, 20_000_000]
    thread_tetap = 4  # Gunakan 4 thread
    
    print("Waktu Eksekusi vs Ukuran Sampel (Thread tetap = 4)")
    print("-" * 50)
    
    for num_samples in ukuran_sampel:
        start_time = time.time()
        pi_est = calculate_pi_parallel(num_samples, thread_tetap)
        end_time = time.time()
        waktu = end_time - start_time
        
        print(f"Sampel: {num_samples:>10,} | Waktu: {waktu:.4f} detik | π: {pi_est:.6f}")
```
