```python
import matplotlib.pyplot as plt

# ========== DATA UNTUK SPEEDUP & EFFICIENCY (Sampel tetap 10M) ==========
process_counts = [1, 2, 4, 8, 16, 32]
waktu_paralel = [3.0945, 2.3734, 0.7773, 0.6126, 0.4632, 0.5371]
waktu_serial = 3.0413  # dari sequential.ipynb

# Hitung speedup
speedups = [waktu_serial / w for w in waktu_paralel]
# Hasil: [0.9828, 1.2814, 3.9126, 4.9640, 6.5658, 5.6624]

# Hitung efficiency
efficiencies = [s / p for s, p in zip(speedups, process_counts)]
# Hasil: [0.9828, 0.6407, 0.9781, 0.6205, 0.4104, 0.1770]

# ========== DATA UNTUK WAKTU VS UKURAN SAMPEL (Thread tetap 4) ==========
ukuran_sampel = [1_000_000, 5_000_000, 10_000_000, 20_000_000]
waktu = [0.1565, 0.4139, 0.7544, 1.5867]

# ========== GRAFIK 1: SPEEDUP VS JUMLAH THREAD ==========
plt.figure(figsize=(10, 6))
plt.plot(process_counts, speedups, 'bo-', linewidth=2, markersize=8, label='Speedup Aktual')
plt.plot(process_counts, process_counts, 'r--', linewidth=2, label='Speedup Ideal')
plt.xlabel('Jumlah Proses')
plt.ylabel('Speedup')
plt.title('Speedup vs Jumlah Proses (Sampel 10 Juta)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('1_speedup.png')
plt.show()

# ========== GRAFIK 2: EFFICIENCY VS JUMLAH THREAD ==========
plt.figure(figsize=(10, 6))
plt.plot(process_counts, efficiencies, 'go-', linewidth=2, markersize=8, label='Efficiency')
plt.axhline(y=1.0, color='r', linestyle='--', label='Efficiency Ideal')
plt.xlabel('Jumlah Proses')
plt.ylabel('Efficiency')
plt.title('Efficiency vs Jumlah Proses (Sampel 10 Juta)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('2_efficiency.png')
plt.show()

# ========== GRAFIK 3: WAKTU VS UKURAN SAMPEL ==========
plt.figure(figsize=(10, 6))
plt.plot(ukuran_sampel, waktu, 'mo-', linewidth=2, markersize=8, label='Waktu (4 Thread)')
plt.xlabel('Ukuran Sampel')
plt.ylabel('Waktu Eksekusi (detik)')
plt.title('Waktu Eksekusi vs Ukuran Sampel (Thread Tetap = 4)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.savefig('3_waktu_vs_sampel.png')
plt.show()
```
