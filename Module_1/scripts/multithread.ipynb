```python
import threading
import random
import time
import math

results = []

def worker(chunk_size):
    inside = 0
    for _ in range(chunk_size):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside += 1
    results.append(inside)

def calculate_pi_thread(num_samples, num_threads):
    chunk = num_samples // num_threads
    threads = []
    results.clear()
    
    for _ in range(num_threads):
        t = threading.Thread(target=worker, args=(chunk,))
        threads.append(t)
        t.start()
    
    for t in threads:
        t.join()
    
    return 4 * sum(results) / num_samples

if __name__ == "__main__":
    num_samples = 10_000_000
    
    for num_threads in [1, 2, 4, 8, 16, 32]:
        start = time.time()
        pi = calculate_pi_thread(num_samples, num_threads)
        end = time.time()
        print(f"Thread: {num_threads} | Waktu: {end-start:.4f} detik | π: {pi:.6f}")
```
